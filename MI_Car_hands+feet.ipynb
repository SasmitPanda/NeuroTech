```python
import mne
from mne.datasets import eegbci

# 1. Configuration
subjects = range(1, 11)  # Subjects 1 through 10
# Runs: 4, 8, 12 (Left/Right Imagery) | 6, 10, 14 (Feet/Tongue Imagery)
runs = [4, 8, 12, 6, 10, 14]

# The 8 most important channels for Motor Imagery (centered on Motor Cortex)
eight_channels = ['FC3', 'FC4', 'C3', 'Cz', 'C4', 'CP3', 'CPz', 'CP4']

all_epochs = []

print("Starting download and preprocessing...")

for sub in subjects:
    # Download files
    files = eegbci.load_data(sub, runs)
    raws = [mne.io.read_raw_edf(f, preload=True) for f in files]
    raw = mne.concatenate_raws(raws)

    # Strip '.' from channel names (e.g., 'C3.' -> 'C3')
    eegbci.standardize(raw)

    # Set standard 10-20 montage
    montage = mne.channels.make_standard_montage('standard_1020')
    raw.set_montage(montage)

    # Pick only our 8 channels
    raw.pick_channels(eight_channels)

    # Find events
    events, _ = mne.events_from_annotations(raw)

    # Define Event IDs based on the dataset documentation:
    # T1 = Left Hand, T2 = Right Hand (Runs 4, 8, 12)
    # T1 = Feet, T2 = Tongue (Runs 6, 10, 14)
    # We will ignore Tongue (T2 in runs 6, 10, 14)
    event_id = {
        'left_hand': 2,  # T1 in hand runs is usually coded as 2
        'right_hand': 3, # T2 in hand runs is usually coded as 3
        'feet': 2        # T1 in feet runs (needs careful mapping based on run index)
    }

    # For simplicity in this initial step, we'll extract all and label manually
    # In a real classifier, you'd separate the 'feet' runs from 'hand' runs.
    # Here is a filtered epoch approach:
    epochs = mne.Epochs(raw, events, event_id=[2, 3], tmin=-1.0, tmax=4.0,
                        baseline=None, preload=True, verbose=False)
    all_epochs.append(epochs)

print(f"Successfully loaded data for {len(subjects)} subjects.")
```

```python
import mne
from mne.datasets import eegbci
import numpy as np

subjects = range(1, 11)
# Hand runs: 4, 8, 12 | Feet runs: 6, 10, 14
hand_runs = [4, 8, 12]
feet_runs = [6, 10, 14]

eight_channels = ['FC3', 'FC4', 'C3', 'Cz', 'C4', 'CP3', 'CPz', 'CP4']
all_epochs_list = []

for sub in subjects:
    print(f"Processing Subject {sub}...")

    # 1. Load Hand and Feet runs separately to label them correctly
    hand_files = eegbci.load_data(sub, hand_runs)
    feet_files = eegbci.load_data(sub, feet_runs)

    raw_hand = mne.concatenate_raws([mne.io.read_raw_edf(f, preload=True) for f in hand_files])
    raw_feet = mne.io.read_raw_edf(feet_files[0], preload=True) # Load first, then append others
    for f in feet_files[1:]:
        raw_feet.append(mne.io.read_raw_edf(f, preload=True))

    for r in [raw_hand, raw_feet]:
        eegbci.standardize(r)
        r.set_montage(mne.channels.make_standard_montage('standard_1020'))
        r.pick_channels(eight_channels)

        # 2. Preprocessing: Filtering
        # Notch filter for line noise (50Hz)
        r.notch_filter(50, fir_design='firwin')
        # Band-pass filter for Motor Imagery (Mu/Beta rhythm: 8-30Hz)
        r.filter(8., 30., fir_design='firwin')

        # 3. Preprocessing: Re-referencing (Common Average Reference)
        r.set_eeg_reference('average')

    # 4. Extract Events and Epoch
    # For Hand runs: T1(2)=Left, T2(3)=Right
    events_hand, _ = mne.events_from_annotations(raw_hand)
    epochs_hand = mne.Epochs(raw_hand, events_hand, event_id={'left': 2, 'right': 3},
                             tmin=1.0, tmax=3.0, baseline=None, preload=True)

    # For Feet runs: T1(2)=Feet, T2(3)=Tongue (We only keep 'feet')
    events_feet, _ = mne.events_from_annotations(raw_feet)
    epochs_feet = mne.Epochs(raw_feet, events_feet, event_id={'feet': 2},
                             tmin=1.0, tmax=3.0, baseline=None, preload=True)

    # Combine epochs for this subject
    sub_epochs = mne.concatenate_epochs([epochs_hand, epochs_feet])
    all_epochs_list.append(sub_epochs)

# Combine all subjects into one massive Epochs object
total_epochs = mne.concatenate_epochs(all_epochs_list)
print("\nFinal Dataset Summary:")
print(total_epochs)
```

```python
from mne.decoding import CSP
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
import numpy as np

# 1. Define the Pipeline
# CSP needs to know how many components (filters) to extract.
# For 8 channels, 4 components is a good starting point.
csp = CSP(n_components=4, reg=None, log=True, norm_trace=False)
lda = LinearDiscriminantAnalysis()

# Create the scikit-learn pipeline
clf = Pipeline([('CSP', csp), ('LDA', lda)])

# 2. Setup Cross-Validation
# 5-fold is standard for small datasets like 10 subjects
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("Subject-wise Classification Accuracies (CSP + LDA):")
print("-" * 40)

subject_scores = []

for i, epochs in enumerate(all_epochs_list):
    # Get the data (X) and the labels (y)
    # X shape: (n_epochs, n_channels, n_times)
    X = epochs.get_data()
    y = epochs.events[:, -1]

    # Run the cross-validation
    # 'cross_val_score' handles the fitting and testing for each fold
    scores = cross_val_score(clf, X, y, cv=cv, n_jobs=1)

    mean_acc = np.mean(scores)
    subject_scores.append(mean_acc)

    print(f"Subject {i+1}: {mean_acc:.2%} (+/- {np.std(scores):.2%})")

# 3. Final Summary
print("-" * 40)
print(f"Overall Mean Accuracy: {np.mean(subject_scores):.2%}")
```
